```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz


import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor,AdaBoostRegressor, GradientBoostingRegressor

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from collections import Counter

file_path = '/Users/f/Documents/Ironhack/ML_Project_Fashion_Trends/salaries _2.csv'

df_main = pd.read_csv(file_path)

print(df_main["employee_residence"].nunique())
print(df_main["company_location"].nunique())
print(df_main["job_title"].nunique())
```

```python
# Define a function to find similar values using fuzzy matching
def find_similar_values(value, values):
    similar_values = []
    for val in values:
        similarity = fuzz.ratio(value, val)
        if similarity > 80:  # Adjust the threshold as needed
            similar_values.append(val)
    return similar_values

# Group similar values and replace them with a common representation
def group_similar_values(df_main, column):
    unique_values = df_main[column].unique()  # Use the correct column name
    grouped_values = {}
    for value in unique_values:
        similar_values = find_similar_values(value, unique_values)
        if similar_values:
            grouped_values[value] = similar_values
    return grouped_values

# Replace similar values with a common representation
grouped_values = group_similar_values(df_main, 'job_title')  
for key, similar_values in grouped_values.items():
    df_main['job_title'].replace(similar_values, key, inplace=True)  

grouped_values_employee_residence = group_similar_values(df_main, 'employee_residence') 
for key, similar_values in grouped_values_employee_residence.items():
    df_main['employee_residence'].replace(similar_values, key, inplace=True)

grouped_values_company_location = group_similar_values(df_main, 'company_location') 
for key, similar_values in grouped_values_company_location.items():
    df_main['company_location'].replace(similar_values, key, inplace=True)



df_main
```

```python
# Splitting the data into features (X) and target (y)
X = df_encoded.drop(columns='salary_in_usd', axis=1)
y = df_encoded['salary_in_usd']
```

### Bagging and Pasting

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Create BaggingRegressor
bagging_reg = BaggingRegressor(base_estimator=lin_reg, n_estimators=10, random_state=42)

# Fit the model on the training data
bagging_reg.fit(X_train_norm, y_train)

# Predict the target variable using the test data
pred_bagging = bagging_reg.predict(X_test_norm)

# Calculate evaluation metrics
mae_bagging = mean_absolute_error(pred_bagging, y_test)
rmse_bagging = mean_squared_error(pred_bagging, y_test, squared=False)
r2_bagging = r2_score(y_test, pred_bagging)

print("Bagging Results:")
print("MAE:", mae_bagging)
print("RMSE:", rmse_bagging)
print("R2 score:", r2_bagging)

# Repeat the process for Pasting
# Create PastingRegressor
pasting_reg = BaggingRegressor(base_estimator=lin_reg, n_estimators=10, random_state=42, bootstrap=False)

# Fit the model on the training data
pasting_reg.fit(X_train_norm, y_train)

# Predict the target variable using the test data
pred_pasting = pasting_reg.predict(X_test_norm)

# Calculate evaluation metrics
mae_pasting = mean_absolute_error(pred_pasting, y_test)
rmse_pasting = mean_squared_error(pred_pasting, y_test, squared=False)
r2_pasting = r2_score(y_test, pred_pasting)

print("\nPasting Results:")
print("MAE:", mae_pasting)
print("RMSE:", rmse_pasting)
print("R2 score:", r2_pasting)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Create DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=42)

# Create BaggingRegressor with DecisionTreeRegressor as base estimator
bagging_tree_reg = BaggingRegressor(base_estimator=tree_reg, n_estimators=10, random_state=42)

# Fit the BaggingRegressor on the training data
bagging_tree_reg.fit(X_train_norm, y_train)

# Predict the target variable using the test data
pred_bagging_tree = bagging_tree_reg.predict(X_test_norm)

# Calculate evaluation metrics for Bagging
mae_bagging_tree = mean_absolute_error(pred_bagging_tree, y_test)
rmse_bagging_tree = mean_squared_error(pred_bagging_tree, y_test, squared=False)
r2_bagging_tree = r2_score(y_test, pred_bagging_tree)

print("Bagging Decision Tree Results:")
print("MAE:", mae_bagging_tree)
print("RMSE:", rmse_bagging_tree)
print("R2 score:", r2_bagging_tree)

# Repeat the process for Pasting
# Create PastingRegressor with DecisionTreeRegressor as base estimator
pasting_tree_reg = BaggingRegressor(base_estimator=tree_reg, n_estimators=10, random_state=42, bootstrap=False)

# Fit the PastingRegressor on the training data
pasting_tree_reg.fit(X_train_norm, y_train)

# Predict the target variable using the test data
pred_pasting_tree = pasting_tree_reg.predict(X_test_norm)

# Calculate evaluation metrics for Pasting
mae_pasting_tree = mean_absolute_error(pred_pasting_tree, y_test)
rmse_pasting_tree = mean_squared_error(pred_pasting_tree, y_test, squared=False)
r2_pasting_tree = r2_score(y_test, pred_pasting_tree)

print("\nPasting Decision Tree Results:")
print("MAE:", mae_pasting_tree)
print("RMSE:", rmse_pasting_tree)
print("R2 score:", r2_pasting_tree)
```

### Ada Boosting

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Create DecisionTreeRegressor as base estimator
base_estimator = DecisionTreeRegressor(max_depth=3)

# Create AdaBoostRegressor with DecisionTreeRegressor as base estimator
adaboost_reg = AdaBoostRegressor(base_estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)

# Fit the AdaBoostRegressor on the training data
adaboost_reg.fit(X_train_norm, y_train)

# Predict the target variable using the test data
pred_adaboost = adaboost_reg.predict(X_test_norm)

# Calculate evaluation metrics for AdaBoost
mae_adaboost = mean_absolute_error(pred_adaboost, y_test)
rmse_adaboost = mean_squared_error(pred_adaboost, y_test, squared=False)
r2_adaboost = r2_score(y_test, pred_adaboost)

print("AdaBoost Results:")
print("MAE:", mae_adaboost)
print("RMSE:", rmse_adaboost)
print("R2 score:", r2_adaboost)
```
